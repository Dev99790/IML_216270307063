```python
'''
Name : Parmar Deva
Enrollment : 216270307063

to sort the DataFrame first by 'name' in 
ascending order
'''

import pandas as pd
import numpy as np
exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James'],
        'score': [12.5, 9, 16.5, np.nan],
        'attempts': [1, 3, 2, 3],
        'qualify': ['yes', 'no', 'yes', 'no']}
labels = ['a', 'b', 'c', 'd']
df = pd.DataFrame(exam_data , index=labels)
print("Orginal rows:")
print(df)
df = df.sort_values(by=['name', 'score'], ascending=[True,False])
print("Sorted data frame :")
print(df)
```
